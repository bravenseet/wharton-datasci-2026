```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Load Data
disparity = pd.read_csv("whl_line_disparity_new.csv")
win_probs = pd.read_csv("Win Probabilities (Logistic Regression).csv")
```

```python
# 2. Extract and average team ratings from win_probs to represent 'Team Strength'
home_ratings = win_probs[['home_team', 'home_rating']].rename(columns={'home_team': 'team', 'home_rating': 'rating'})
away_ratings = win_probs[['away_team', 'away_rating']].rename(columns={'away_team': 'team', 'away_rating': 'rating'})
team_ratings = pd.concat([home_ratings, away_ratings]).groupby('team').mean().reset_index()

# 3. Merge Datasets
merged = pd.merge(disparity, team_ratings, on='team')
```

```python
# 4. Set Plot Style
plt.figure(figsize=(10, 6), dpi=300)
sns.set_theme(style="whitegrid", context="talk")

# 5. Create Scatter Plot with Trendline
ax = sns.regplot(
    data=merged, 
    y='disparity_log_ratio', 
    x='rating', 
    scatter_kws={'alpha': 0.6, 'color': '#2C3E50', 'edgecolor': 'w', 's': 70},
    line_kws={'color': '#E74C3C', 'linewidth': 3}
)

# 6. Customize Labels and Titles
plt.title("Does a Balanced Attack Matter?\n", fontsize=18, weight='bold', loc='left')
#plt.suptitle("No clear link found between offensive line disparity and overall team strength", fontsize=13, x=0.125, y=0.90, ha='left')

plt.ylabel("Log Line Disparity Ratio\nabs(ln(1st Line xG/60 ÷ 2nd Line xG/60))\n← More Evenly Matched | Top-Heavy →", fontsize=12)
plt.xlabel("Team Strength (Elo Rating)", fontsize=12)

# 7. Add Caption
plt.figtext(0.90, 0.01, "Source: WHL Line Disparity & Win Probabilities Data", ha="right", fontsize=9, color="gray", style="italic")

# 8. Clean up and Save
sns.despine(left=True, bottom=True)
plt.tight_layout(rect=[0, 0.05, 1, 0.92]) 
#plt.show()

file_name = "phase1c_disparityvssuccess.png"
plt.savefig(file_name, bbox_inches="tight")
plt.close()
```
